# EDA

```python
# create a function to show missing values and summary statistics
def eda(df):
    eda_df = {}
    eda_df["null_sum"] = df.isna().sum()
    eda_df["null_pct"] = df.isna().mean()
    eda_df["dtypes"] = df.dtypes
    eda_df["count"] = df.count()
    eda_df["mean"] = df.mean()
    eda_df["median"] = df.median()
    eda_df["min"] = df.min()
    eda_df["max"] = df.max()
    
    return pd.DataFrame(eda_df)
```

```python
eda(df)
```
